# 05 Undersampling benchmark

Controlled synthetic undersampling notebook for baseline vs placeholder QCSM comparison.

```python
import numpy as np
from qsg_mri.metrics import nmse
from qsg_mri.phantoms import synthetic_multicoil_sensitivities, synthetic_phantom
from qsg_mri.reconstruction import baseline_reconstruction, qcsm_reconstruction_placeholder

size = 64
phantom = synthetic_phantom(size=size)
sens = synthetic_multicoil_sensitivities(num_coils=6, size=size)
coil_images = sens * phantom[None, :, :]
kspace = np.fft.ifftshift(np.fft.fft2(np.fft.fftshift(coil_images, axes=(-2, -1)), axes=(-2, -1)), axes=(-2, -1))
mask = np.zeros((size, size), dtype=float)
mask[::4, :] = 1.0
mask[size//2-6:size//2+6, :] = 1.0
kspace_u = kspace * mask[None, :, :]
baseline = baseline_reconstruction(kspace_u)
q_recon = qcsm_reconstruction_placeholder(kspace_u)['reconstruction']
print('baseline NMSE', nmse(phantom, baseline))
print('qcsm NMSE', nmse(phantom, q_recon))
```
