# 01 MRI k-space baseline

Synthetic, research-only walkthrough of standard complex MRI baseline reconstruction.

```python
import numpy as np
from qsg_mri.phantoms import synthetic_phantom

phantom = synthetic_phantom(size=64)
kspace = np.fft.ifftshift(np.fft.fft2(np.fft.fftshift(phantom)))
recon = np.fft.ifftshift(np.fft.ifft2(np.fft.fftshift(kspace))).real
print(phantom.shape, recon.shape)
```
