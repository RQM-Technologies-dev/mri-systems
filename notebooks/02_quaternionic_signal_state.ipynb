# 02 Quaternionic signal state

Construct fixed-axis and general-axis quaternionic states for synthetic MRI examples.

```python
import numpy as np
from qsg_mri.quaternion import fixed_axis_complex_state, from_axis_angle

q_fixed = fixed_axis_complex_state(amplitude=1.0, phi=np.pi / 3)
q_general = from_axis_angle(amplitude=1.0, phi=np.pi / 3, axis=np.array([0.0, 1.0, 0.0]))
print('fixed:', q_fixed)
print('general:', q_general)
```
