# 04 Coherence defect map

Inspect coherence-defect behavior from placeholder QCSM outputs in synthetic settings.

```python
import numpy as np
from qsg_mri.phantoms import synthetic_multicoil_sensitivities, synthetic_phantom
from qsg_mri.reconstruction import qcsm_reconstruction_placeholder

phantom = synthetic_phantom(size=64)
sens = synthetic_multicoil_sensitivities(num_coils=8, size=64)
coil_images = sens * phantom[None, :, :]
kspace = np.fft.ifftshift(np.fft.fft2(np.fft.fftshift(coil_images, axes=(-2, -1)), axes=(-2, -1)), axes=(-2, -1))
coherence_defect = qcsm_reconstruction_placeholder(kspace)['coherence_defect']
print(float(np.min(coherence_defect)), float(np.max(coherence_defect)))
```
